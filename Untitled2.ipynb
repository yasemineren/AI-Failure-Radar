```python
import os
import matplotlib.pyplot as plt
import cv2
import random
import numpy as np

# Veri setinin ana klasörünü bulalım
# (Genelde 'Dataset_BUSI_with_GT' adında çıkar)
base_dir = "Dataset_BUSI_with_GT"

# Kanserli (Malignant) klasörüne bakalım
img_dir = os.path.join(base_dir, "malignant")

# Klasördeki tüm resim dosyalarını listele
all_files = os.listdir(img_dir)

# Sadece orijinal resimleri al (sonu 'mask.png' ile bitmeyenler)
img_files = [f for f in all_files if not f.endswith('mask.png')]

# Rastgele bir hasta seç
random_img = random.choice(img_files)

# Dosya yollarını oluştur
img_path = os.path.join(img_dir, random_img)
# Maske dosyası genelde 'resimadi_mask.png' formatındadır
mask_path = os.path.join(img_dir, random_img.replace(".png", "_mask.png"))

# Görüntüleri oku
img = cv2.imread(img_path, 0) # 0 = Gri tonlamalı oku
mask = cv2.imread(mask_path, 0)

# Görselleştirme
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title(f"Ham Ultrason Görüntüsü\n({random_img})")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(mask, cmap='gray')
plt.title("Doktorun İşaretlediği Kanser (Maske)")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img, cmap='gray')
plt.imshow(mask, cmap='jet', alpha=0.3) # Üst üste bindir
plt.title("Çakıştırılmış Görüntü (Overlay)")
plt.axis('off')

plt.tight_layout()
plt.show()

print(f"Görüntü Boyutu: {img.shape}")
print(f"Maske Boyutu: {mask.shape}")
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
from sklearn.model_selection import train_test_split

# 1. AYARLAR (Konfigürasyon)
IMG_SIZE = 224
BATCH_SIZE = 32 # Modele her seferinde 32 resim birden göndereceğiz

# 2. ÖZEL VERİ SETİ SINIFI (İşte Mühendislik Burada!)
class BreastCancerDataset(Dataset):
    def __init__(self, file_list, labels, transform=None):
        self.file_list = file_list
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.file_list)

    def __getitem__(self, idx):
        # Görüntüyü diskten oku
        img_path = self.file_list[idx]
        image = Image.open(img_path).convert("RGB") # Renkli okuyoruz (Transfer Learning için gerekli)

        # Etiketi al (0 veya 1)
        label = self.labels[idx]

        # Dönüşümleri uygula (Resize, Tensor vb.)
        if self.transform:
            image = self.transform(image)

        # PyTorch float tensor istiyor, etiketi de tensor yapalım
        return image, torch.tensor(label, dtype=torch.float32)

# 3. DOSYA YOLLARINI VE ETİKETLERİ HAZIRLAMA
all_files = []
all_labels = []

# Klasörleri tanımla: 0 = Temiz/İyi Huylu, 1 = Kötü Huylu (Kanser)
# (Klasör isimleri indirdiğimiz veri setine göre ayarlandı)
categories = {
    "normal": 0,
    "benign": 0,
    "malignant": 1
}

base_dir = "Dataset_BUSI_with_GT"

print("Dosyalar taranıyor...")
for category, label in categories.items():
    folder_path = os.path.join(base_dir, category)
    # Klasör var mı diye kontrol et (Bazen isimler farklı olabilir)
    if os.path.exists(folder_path):
        for fname in os.listdir(folder_path):
            if not fname.endswith("mask.png"): # Maskeleri alma, sadece orijinal resimler
                all_files.append(os.path.join(folder_path, fname))
                all_labels.append(label)

print(f"Toplam Görüntü Sayısı: {len(all_files)}")

# 4. VERİYİ BÖLME (Eğitim ve Test)
# %80 Eğitim (Train), %20 Test (Validation)
X_train, X_val, y_train, y_val = train_test_split(all_files, all_labels, test_size=0.2, random_state=42)

# 5. DÖNÜŞÜMLERİ TANIMLA
# ResNet için standartlaştırma değerleri (ImageNet istatistikleri)
transform_pipeline = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# 6. DATA LOADERLARI OLUŞTUR
train_dataset = BreastCancerDataset(X_train, y_train, transform=transform_pipeline)
val_dataset = BreastCancerDataset(X_val, y_val, transform=transform_pipeline)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"Eğitim Seti: {len(train_dataset)} görüntü")
print(f"Test Seti: {len(val_dataset)} görüntü")
print("✅ Veri hattı kuruldu! Motor çalışmaya hazır.")
```

```python
import torchvision.models as models
import torch.nn as nn
import torch.optim as optim

# 1. CİHAZ SEÇİMİ (GPU varsa kullan, yoksa CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Eğitim şu cihazda yapılacak: {device}")

# 2. MODELİ HAZIRLA (ResNet18)
# 'weights=ResNet18_Weights.DEFAULT' en güncel ağırlıkları indirir
model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

# Modelin son katmanını (fc) değiştiriyoruz.
# ResNet18'in orijinal çıkışı 1000 sınıftır. Biz bunu 1'e indiriyoruz (0 veya 1 tahmini için).
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 1)

# Modeli GPU'ya taşı
model = model.to(device)

# 3. EĞİTİM KURALLARI (Hyperparameters)
# Hata Hesaplayıcı: Binary Cross Entropy (İkili sınıflandırma için standart)
criterion = nn.BCEWithLogitsLoss()

# Öğretmen (Optimizer): Adam (Hızlı ve akıllı öğrenir)
optimizer = optim.Adam(model.parameters(), lr=0.0001) # Düşük öğrenme hızı (Fine-tuning)

# 4. EĞİTİM DÖNGÜSÜ (Training Loop)
EPOCHS = 5 # Şimdilik 5 tur döneceğiz (Demo olduğu için kısa tutuyoruz)

print("🚀 Eğitim Başlıyor...")

for epoch in range(EPOCHS):
    model.train() # Modeli eğitim moduna al
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        # Verileri GPU'ya gönder
        images = images.to(device)
        labels = labels.to(device).unsqueeze(1) # Boyut düzeltme (Batch, 1) yapıyoruz

        # a. Sıfırla (Önceki turun türevlerini temizle)
        optimizer.zero_grad()

        # b. Tahmin Et (Forward Pass)
        outputs = model(images)

        # c. Hatayı Hesapla
        loss = criterion(outputs, labels)

        # d. Geriye Yayılım (Backpropagation - Öğrenme burada olur!)
        loss.backward()

        # e. Ağırlıkları Güncelle
        optimizer.step()

        # İstatistikleri topla
        running_loss += loss.item()

        # Doğruluk (Accuracy) hesabı (Logit > 0 ise 1 kabul et)
        predicted = (outputs > 0).float()
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    # Her epoch sonunda rapor ver
    epoch_loss = running_loss / len(train_loader)
    epoch_acc = 100 * correct / total
    print(f"Epoch [{epoch+1}/{EPOCHS}] -> Loss: {epoch_loss:.4f} | Accuracy: %{epoch_acc:.2f}")

print("✅ Eğitim Tamamlandı! Modelin artık ultrason okumayı öğrendi.")
```

```python
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget
from pytorch_grad_cam.utils.image import show_cam_on_image
import cv2
import numpy as np

# 1. Grad-CAM Ayarları
# ResNet'in son evrişim katmanına (layer4) kanca atıyoruz
target_layers = [model.layer4[-1]]
cam = GradCAM(model=model, target_layers=target_layers)

# 2. Test Setinden Rastgele Kanserli Bir Resim Seçelim
model.eval() # Modeli test moduna al
idx = 0
found = False

# Sadece 'Malignant' (Kanserli) olan bir örnek bulana kadar dön
while not found:
    idx = np.random.randint(0, len(val_dataset))
    img_tensor, label = val_dataset[idx]
    if label.item() == 1.0: # 1 = Kanser
        found = True

# 3. Hazırlık
input_tensor = img_tensor.unsqueeze(0).to(device) # (1, 3, 224, 224)

# 4. Grad-CAM Çalıştır (Sihirli Kısım)
# Modelin 'Kanser' (1) sınıfı için nereye baktığını hesapla
targets = [ClassifierOutputTarget(0)] # Logitlerde tek çıkış var, o yüzden 0. index
grayscale_cam = cam(input_tensor=input_tensor, targets=targets)
grayscale_cam = grayscale_cam[0, :] # İlk örneği al

# 5. Görselleştirme
# Orijinal resmi geri al (Normalizasyonu tersine çevirip görsel hale getir)
rgb_img = img_tensor.permute(1, 2, 0).cpu().numpy()
rgb_img = (rgb_img - rgb_img.min()) / (rgb_img.max() - rgb_img.min()) # 0-1 arasına çek

# Isı haritasını orijinal resmin üzerine bindir
visualization = show_cam_on_image(rgb_img, grayscale_cam, use_rgb=True)

# Yan yana çizelim
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(rgb_img)
plt.title("Orijinal Ultrason Görüntüsü")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(visualization)
plt.title(f"Modelin Odaklandığı Yer (Grad-CAM)\nGerçek Etiket: Kanser")
plt.axis('off')

plt.show()
```
